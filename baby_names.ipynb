```python
# import packages
from textgenrnn import textgenrnn
import sys
```

```python
# create rnn to generate baby names
textgen = textgenrnn(name="baby_names")
textgen.reset()
```

```python
# training based on 2018 list of names associated with Social Security Number
# original data can be found at https://catalog.data.gov/dataset/baby-names-from-social-security-card-applications-national-level-data
textgen.train_from_file('set_names.txt',
                        new_model=True,
                        rnn_bidirectional=True,
                        rnn_size=64,
                        dim_embeddings=300,
                        num_epochs=10)
```

```python
# generate 25 samples with varying temperatures (higher the temperature, the more "creative")
textgen.generate_samples(n=25, temperatures=[0.2, 0.5, 1.0, 1.2, 1.5, 2.0])
```
